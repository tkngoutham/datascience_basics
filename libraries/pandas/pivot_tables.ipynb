<br/>
<b style="color:blue;">Below Pivot Table gives us the mean of weights in both Males and Females
across all the countries. We have the countries as the columns and gender as the index.
While the inner values of the Table are the sum of weights </b>

```python
# Constructing the Pivot Table

table = pd.pivot_table(df, values='weight', index=['gender'], columns=['Country'], aggfunc=np.mean, fill_value=0)
```
